```python
import os
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.models import load_model

model = load_model(r"C:\Users\ADMIN\Music\chest_xray\chest_xray_project_clean\model.h5")

folder = r"C:\Users\ADMIN\Music\chest_xray\chest_xray_project_clean"

classes = ["Normal", "Pneumonia"]

for filename in os.listdir(folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img_path = os.path.join(folder, filename)

        img = image.load_img(img_path, target_size=(224,224))
        img_array = image.img_to_array(img)
        img_array = img_array/255.0
        img_array = np.expand_dims(img_array, axis=0)

        pred = model.predict(img_array)
        predicted_class = np.argmax(pred)

        print(filename, " ---> ", classes[predicted_class])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model

model = load_model("model.h5")

test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(
    r"C:\Users\ADMIN\Music\chest_xray\chest_xray_project_clean\test",
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical"
)

loss, acc = model.evaluate(test_data)

print("Test Accuracy:", acc)
print("Test Loss:", loss)
```
